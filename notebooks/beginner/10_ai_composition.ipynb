## 💡 Key Concept: What is a Large Language Model (LLM)?

An **LLM** is an AI system trained on massive text data that can generate
human-like text, answer questions, and write code.

How it works (simplified):
1. Trained on billions of words (books, code, docs)
2. Learns: "given this input, what text likely follows?"
3. Generates responses word by word

Examples: ChatGPT, Claude, Llama, Gemini

## 💡 Key Concept: Heuristic vs. LLM Mode

| | Heuristic Mode | LLM Mode |
|---|---|---|
| How | Rules + templates | AI generation |
| Deterministic? | ✅ Always same output | ❌ May vary |
| Offline? | ✅ No internet needed | ❌ Needs API |
| Cost | Free | ~$0.01-0.10/call |
| Best for | CI/CD, reproducible | Creative, complex |

Scalable always works in heuristic mode. LLM is an optional enhancement.

````python
# Write a base manifest for the AI tools to work with
manifest_content = """\
version: 1
project:
  name: water-resources

targets:
  local:
    provider: local
    max_workers: 4
    threads_per_worker: 1
    processes: false
    containers: none

components:
  hydrology:
    cpus: 4
    memory: 16G

tasks:
  run_hydrology:
    component: hydrology
"""

with open("scalable.yaml", "w") as f:
    f.write(manifest_content)
print("Base manifest ready for AI assistant demos.")
````

````python
# Demonstrate AI component onboarding (heuristic mode)
import os
from scalable.ai.component_onboarding import onboard_component

# Create a mock model directory for onboarding
os.makedirs("watershed_model", exist_ok=True)
with open("watershed_model/README.md", "w") as f:
    f.write("# WaterShed Model\nHydrological watershed model for runoff simulation\n")

# Explanation: onboard_component analyzes a directory and generates
# a component definition from what it finds
result = onboard_component(
    "./watershed_model",
    name="watershed",
    no_ai=True,  # Use heuristic mode (no LLM needed)
)

print("Generated component definition (heuristic mode):")
print("─" * 50)
print(result.component_yaml)
````

## 💡 What Just Happened?

The `onboard_component` assistant:
1. Took your high-level inputs (name, image, resources)
2. Applied templates with sensible defaults
3. Inferred tags from the description
4. Generated matching task bindings

In **LLM mode**, it could also suggest optimal allocations,
recommend mount points, and generate preload scripts.

````python
# Demonstrate plan explanation (heuristic mode)
from scalable.ai.plan_explain import explain_plan

# Create a minimal plan dict for explanation
plan_data = {
    "target_name": "local",
    "provider": "local",
    "manifest_lock": "sha256:abc123",
    "scale_plan": {"hydrology": {"workers": 4, "cpus": 4, "memory": "16G"}},
}

# Explanation: explain_plan generates human-readable explanation
explanation = explain_plan(
    plan_data=plan_data,
    no_ai=True,
)

print("Plan Explanation:")
print("═" * 50)
print(explanation.render_text())
````

## 💡 Key Concept: Human-in-the-Loop

**Human-in-the-loop** = AI suggests, human decides.

AI-generated output should always be:
1. **Validated**: `scalable validate` on generated manifests
2. **Reviewed**: Check resource allocations, security settings
3. **Tested**: `--dry-run` before real deployment
4. **Approved**: Human confirms before applying

Never blindly trust AI output — it can generate plausible-looking
but incorrect configurations.

## 💡 Key Concept: Templates (Jinja2)

**Templates** = pre-structured documents with fill-in-the-blank placeholders:

```
components:
  {{ name }}:
    cpus: {{ cpus }}
    memory: {{ memory }}
```

Heuristic mode uses templates extensively → predictable, fast, reproducible.
